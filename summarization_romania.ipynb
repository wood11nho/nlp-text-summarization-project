```python
import os
from datasets import load_dataset
```

```python
# Create a folder for the news articles and summaries

ro_dataset = load_dataset("readerbench/ro-text-summarization")
ro_dataset["train"][5]
# get total number of ro_dataset

number_of_articles = len(ro_dataset["train"])
for line in ro_dataset["train"]:
    category = line["Category"]
    content = line["Content"]
    summary = line["Summary"]
    if category not in os.listdir("romanian_news/news_articles"):
        os.makedirs(f"romanian_news/news_articles/{category}")
    if category not in os.listdir("romanian_news/summaries"):
        os.makedirs(f"romanian_news/summaries/{category}")
    # create a file for content in news_articles and summary in summaries, name it 000...n.txt, and get the number of digits we need from number_of_articles
    txt_name_format = f"{{:0{len(str(number_of_articles))}d}}.txt"
    i = len(os.listdir(f"romanian_news/news_articles/{category}")) + 1
    with open(f"romanian_news/news_articles/{category}/{txt_name_format.format(i)}", "w", encoding="utf-8") as file:
        file.write(content)
    j = len(os.listdir(f"romanian_news/summaries/{category}")) + 1
    with open(f"romanian_news/summaries/{category}/{txt_name_format.format(j)}", "w", encoding="utf-8") as file:
        file.write(summary)
```

UNIVERSAL SENTENCE ENCODER

```python
embed = hub.load("https://tfhub.dev/google/universal-sentence-encoder/4")

def sentence_similarity_use(sentence1, sentence2, embed):
    x = embed([sentence1])[0]
    y = embed([sentence2])[0]
    return 1 - (np.dot(x, y) / (np.linalg.norm(x) * np.linalg.norm(y)))
```

```python
print("sentence 1: ", mod_sent[1])
print("sentence 2: ", mod_sent[2])

print("Similarity: ", sentence_similarity_use(mod_sent[1], mod_sent[2], embed))
```

```python
def similarity_matrix_use(sentences, embed):
    n = len(sentences)
    matrix = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            matrix[i][j] = sentence_similarity_use(sentences[i], sentences[j], embed)
    return matrix

sim_matrix_use = similarity_matrix_use(mod_sent, embed)
sim_matrix_use
```

```python
from bokeh.io import output_notebook, show, save
from bokeh.models import Range1d, Circle, ColumnDataSource, MultiLine, HoverTool
from bokeh.plotting import figure, from_networkx

import networkx as nx
output_notebook()

g = nx.Graph()

for i in range(sim_matrix_use.shape[0]):
    for j in range(sim_matrix_use.shape[1]):
        if i != j and sim_matrix_use[i][j] > 0.5:
            g.add_edge(i, j, weight=sim_matrix_use[i][j])
            
HOVER_TOOLTIPS = [("sent_tok", "@index")]
plot = figure(title="Sentence Similarity Graph", x_range = Range1d(-10.1, 10.1), y_range = Range1d(-10.1, 10.1),
                tools="pan,wheel_zoom,save,reset,hover", tooltips=HOVER_TOOLTIPS, active_scroll="wheel_zoom")

network_graph = from_networkx(g, nx.spring_layout, scale=7, center=(0, 0))
network_graph.node_renderer.glyph = Circle(radius=0.1, fill_color='skyblue')
network_graph.edge_renderer.glyph = MultiLine(line_alpha=0.8, line_width=1)
plot.renderers.append(network_graph)

show(plot)
```

```python
def generate_summary_use(text, top_n_sentences, embeds):
    summarize_text = []
    sentences = read_article(text)
    sentence_similarity_martix = similarity_matrix_use(sentences, embeds)
    sentence_similarity_graph = nx.from_numpy_array(sentence_similarity_martix)
    scores = nx.pagerank(sentence_similarity_graph)
    ranked_sentences = sorted(((scores[i], s) for i, s in enumerate(sentences)), reverse=True)
    
    for i in range(top_n_sentences):
        summarize_text.append("".join(ranked_sentences[i][1]))
        if len(" ".join(summarize_text)) > len(summary_text):
            break
    return " ".join(summarize_text)

summary_generated = generate_summary_use(text, 3, embed)
print("Generated summary: ", summary_generated)
print("Original summary: ", summary_text)
```

SENTENCE TRANSFORMERS

```python
from sentence_transformers import SentenceTransformer
import scipy.spatial

model = SentenceTransformer('paraphrase-xlm-r-multilingual-v1')

def sentence_similarity_st(sentence1, sentence2, model):
    embeddings = model.encode([sentence1, sentence2])
    cosine_sim = 1 - scipy.spatial.distance.cosine(embeddings[0], embeddings[1])
    return cosine_sim

# Example usage
print("sentence 1: ", mod_sent[1])
print("sentence 2: ", mod_sent[2])

print("Similarity: ", sentence_similarity_st(mod_sent[1], mod_sent[2], model))
```

```python
def similarity_matrix_st(sentences, model):
    n = len(sentences)
    matrix = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            matrix[i][j] = sentence_similarity_st(sentences[i], sentences[j], model)
    return matrix

sim_matrix_st = similarity_matrix_st(mod_sent, model)
sim_matrix_st
```

```python
def generate_summaries_and_stats(df, model_use, model_st, top_n_articles):
    bleu_scores_use, bleu_scores_st = [], []
    sim_scores_use, sim_scores_st = [], []
    
    with open('generated_summaries.csv', 'w', encoding='utf-8') as file:
        file.write("Text\tGenerated Summary (USE)\tGenerated Summary (ST)\tOriginal Summary\n")
    
    for idx in range(top_n_articles):
        text_path = df.iloc[idx]['path']
        with open(text_path, 'r', encoding='utf-8') as file:
            text = file.read()
        summary_text = read_summary_from_path(df, 'article_or_summary', 'summaries', idx)
        
        # Generate summaries
        summary_generated_use = generate_summary_use(text, 3, model_use)
        summary_generated_st = generate_summary_st(text, 3, model_st)
        
        # Add the text, summary_generated_use, summary_generated_st and summary_text to the D:\Facultate\Anul 3\Sem 2\NLP\Proiect\romanian_news\generated_summaries.csv
        with open('generated_summaries.csv', 'a', encoding='utf-8') as file:
            file.write(f"{text}\t{summary_generated_use}\t{summary_generated_st}\t{summary_text}\n")
        
        # Compute BLEU scores
        bleu_use = bleu_score(summary_generated_use, summary_text)
        bleu_st = bleu_score(summary_generated_st, summary_text)
        bleu_scores_use.append(bleu_use)
        bleu_scores_st.append(bleu_st)
        
        # Compute similarity scores
        sim_use = sentence_similarity_use(summary_generated_use, summary_text, model_use)
        sim_st = sentence_similarity_st(summary_generated_st, summary_text, model_st)
        sim_scores_use.append(sim_use)
        sim_scores_st.append(sim_st)
    
    return bleu_scores_use, bleu_scores_st, sim_scores_use, sim_scores_st
```

```python
from bokeh.layouts import gridplot

def plot_statistics(bleu_scores_use, bleu_scores_st, sim_scores_use, sim_scores_st):
    x = list(range(1, top_n_articles + 1))  # Convert range to list here

    # Prepare data for plotting
    source = ColumnDataSource(data=dict(x=x, bleu_use=bleu_scores_use, bleu_st=bleu_scores_st,
                                        sim_use=sim_scores_use, sim_st=sim_scores_st))
    
    # Create plots
    p1 = figure(title="BLEU Scores Comparison", x_axis_label='Article', y_axis_label='BLEU Score', tools="")
    p1.line('x', 'bleu_use', source=source, legend_label="USE", line_width=2, line_color="blue")
    p1.line('x', 'bleu_st', source=source, legend_label="ST", line_width=2, line_color="green")
    
    p2 = figure(title="Sentence Similarity Comparison", x_axis_label='Article', y_axis_label='Similarity Score', tools="")
    p2.line('x', 'sim_use', source=source, legend_label="USE", line_width=2, line_color="blue")
    p2.line('x', 'sim_st', source=source, legend_label="ST", line_width=2, line_color="green")
    
    # Show plots
    p = gridplot([[p1, p2]])
    show(p)

# Example usage
bleu_scores_use, bleu_scores_st, sim_scores_use, sim_scores_st = generate_summaries_and_stats(df, embed, model, top_n_articles)
plot_statistics(bleu_scores_use, bleu_scores_st, sim_scores_use, sim_scores_st)
```
